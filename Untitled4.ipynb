```python
import os

def count_images_in_dataset(root_dir):
    stats = {}
    for sensor in os.listdir(root_dir):
        sensor_path = os.path.join(root_dir, sensor)
        if not os.path.isdir(sensor_path):
            continue
        stats[sensor] = {'Live': 0, 'Fake': 0, 'materials': {}}
        for live_fake in ['Live', 'Fake']:
            path = os.path.join(sensor_path, live_fake)
            if not os.path.exists(path):
                continue
            if live_fake == 'Live':
                count = len(os.listdir(path))
                stats[sensor]['Live'] = count
            else:
                total_fake = 0
                for material in os.listdir(path):
                    mat_path = os.path.join(path, material)
                    count = len(os.listdir(mat_path))
                    stats[sensor]['materials'][material] = count
                    total_fake += count
                stats[sensor]['Fake'] = total_fake
    return stats

training_stats = count_images_in_dataset("LivDet2015/Training")
testing_stats = count_images_in_dataset("LivDet2015/Testing")

from pprint import pprint
print("Training Stats:")
pprint(training_stats)
print("\nTesting Stats:")
pprint(testing_stats)
```

```python
import matplotlib.pyplot as plt

def plot_live_vs_fake(stats, title):
    sensors = list(stats.keys())
    live_counts = [stats[s]['Live'] for s in sensors]
    fake_counts = [stats[s]['Fake'] for s in sensors]

    x = range(len(sensors))
    plt.figure(figsize=(10, 6))
    plt.bar(x, live_counts, width=0.4, label='Live', align='center')
    plt.bar([p + 0.4 for p in x], fake_counts, width=0.4, label='Fake', align='center')
    plt.xticks([p + 0.2 for p in x], sensors)
    plt.ylabel("Number of Images")
    plt.title(title)
    plt.legend()
    plt.show()

plot_live_vs_fake(training_stats, "Training Set - Live vs Fake")
plot_live_vs_fake(testing_stats, "Testing Set - Live vs Fake")
```

```python
import pandas as pd

def get_material_distribution(stats):
    rows = []
    for sensor, info in stats.items():
        for material, count in info['materials'].items():
            rows.append({'Sensor': sensor, 'Material': material, 'Count': count})
    return pd.DataFrame(rows)

df_train = get_material_distribution(training_stats)
df_test = get_material_distribution(testing_stats)

import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(data=df_train, x="Sensor", y="Count", hue="Material")
plt.title("Training Set - Fake Materials Distribution")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_test, x="Sensor", y="Count", hue="Material")
plt.title("Testing Set - Fake Materials Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
import os
from PIL import Image, UnidentifiedImageError
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import torch.optim as optim
import timm
from tqdm import tqdm

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 50
LR = 1e-4
PATIENCE = 3
SAVE_EVERY_N_EPOCHS = 5  

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_DIR = "LivDet2015/Training"
SENSOR_NAME = "GreenBit"
CHECKPOINT_DIR = "checkpoints"
os.makedirs(CHECKPOINT_DIR, exist_ok=True)

# Transformations des images
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# Dataset personnalis
class LivDetBinaryDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        sensor_path = os.path.join(root_dir, SENSOR_NAME)
        for label in ['Live', 'Fake']:
            label_path = os.path.join(sensor_path, label)
            if not os.path.isdir(label_path):
                continue
            if label == 'Live':
                for img_file in os.listdir(label_path):
                    img_path = os.path.join(label_path, img_file)
                    if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                        self.samples.append((img_path, 0))
            else:
                for material in os.listdir(label_path):
                    material_path = os.path.join(label_path, material)
                    if not os.path.isdir(material_path):
                        continue
                    for img_file in os.listdir(material_path):
                        img_path = os.path.join(material_path, img_file)
                        if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                            self.samples.append((img_path, 1))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            with Image.open(img_path) as img:
                img = img.convert('L')
                if self.transform:
                    img = self.transform(img)
                return img, label
        except (OSError, UnidentifiedImageError) as e:
            print(f"[Skipping] Image corrompue ou illisible: {img_path} ({e})")
            return self.__getitem__((idx + 1) % len(self.samples))  # skip vers l'image suivante

# EarlyStopping
class EarlyStopping:
    def __init__(self, patience=5):
        self.patience = patience
        self.best_loss = float('inf')
        self.counter = 0
        self.early_stop = False

    def step(self, val_loss):
        if val_loss < self.best_loss:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            print(f"  EarlyStopping counter: {self.counter}/{self.patience}")
            if self.counter >= self.patience:
                self.early_stop = True

# Charge les donnes et cre les DataLoader
dataset = LivDetBinaryDataset(DATA_DIR, transform=transform)
total_size = len(dataset)
val_size = int(0.2 * total_size)  # 20% pour validation
train_size = total_size - val_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Modle et initialisation
model = timm.create_model('convnext_tiny', pretrained=True, num_classes=2)
model.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LR)

early_stopper = EarlyStopping(patience=PATIENCE)

# Enregistrement du meilleur modle
best_val_loss = float('inf')

# Boucle d'entranement et validation
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    # Entranement
    loop = tqdm(train_loader, desc=f"Epoch [{epoch+1}/{EPOCHS}]")
    for inputs, labels in loop:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        loop.set_postfix(loss=loss.item(), acc=100. * correct / total)

    avg_train_loss = total_loss / len(train_loader)

    # Validation
    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    avg_val_loss = val_loss / len(val_loader)
    val_acc = 100. * val_correct / val_total

    print(f"Train Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_acc:.2f}%")

    # Enregistrer le modle si la validation loss s'amliore
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        checkpoint = {
            'epoch': epoch + 1,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'train_loss': avg_train_loss,
            'val_loss': avg_val_loss
        }
        checkpoint_path = os.path.join(CHECKPOINT_DIR, f'checkpoint_epoch_{epoch+1}.pth')
        torch.save(checkpoint, checkpoint_path)
        print(f"   Checkpoint saved: {checkpoint_path}")

    # Early stopping
    early_stopper.step(avg_val_loss)
    if early_stopper.early_stop:
        print("Early stopping triggered.")
        break

print("Training complete.")
```

```python
#
import os
import numpy as np
from PIL import Image, UnidentifiedImageError
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import timm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 32
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_DIR = "LivDet2015/Testing"
SENSOR_NAME = "CrossMatch"

# Transformations des images
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# Dataset personnalis
class LivDetBinaryDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        sensor_path = os.path.join(root_dir, SENSOR_NAME)
        for label in ['Live', 'Fake']:
            label_path = os.path.join(sensor_path, label)
            if not os.path.isdir(label_path):
                continue
            if label == 'Live':
                for img_file in os.listdir(label_path):
                    img_path = os.path.join(label_path, img_file)
                    if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                        self.samples.append((img_path, 0))
            else:
                for material in os.listdir(label_path):
                    material_path = os.path.join(label_path, material)
                    if not os.path.isdir(material_path):
                        continue
                    for img_file in os.listdir(material_path):
                        img_path = os.path.join(material_path, img_file)
                        if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                            self.samples.append((img_path, 1))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            with Image.open(img_path) as img:
                img = img.convert('L')
                if self.transform:
                    img = self.transform(img)
                return img, label
        except (OSError, UnidentifiedImageError) as e:
            print(f"[Skipping] Image corrompue ou illisible: {img_path} ({e})")
            return self.__getitem__((idx + 1) % len(self.samples))  # skip vers l'image suivante

# Charge les donnes de test
test_dataset = LivDetBinaryDataset(DATA_DIR, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Charger le modle pralablement sauvegard
CHECKPOINT_PATH = "checkpoints/CrossMatch/checkpoint_epoch_2.pth" 
checkpoint = torch.load(CHECKPOINT_PATH)
model = timm.create_model('convnext_tiny', pretrained=False, num_classes=2)
model.load_state_dict(checkpoint['model_state_dict'])
model.to(DEVICE)
model.eval()

# Variables pour le calcul des mtriques
all_labels = []
all_preds = []

# Test du modle
test_loss = 0
criterion = nn.CrossEntropyLoss()

with torch.no_grad():
    loop = tqdm(test_loader, desc="Testing")
    for inputs, labels in loop:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        test_loss += loss.item()

        _, preds = torch.max(outputs, 1)
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

        loop.set_postfix(loss=test_loss / len(test_loader))

# Calcul des mtriques
avg_test_loss = test_loss / len(test_loader)
test_acc = accuracy_score(all_labels, all_preds)
test_precision = precision_score(all_labels, all_preds)
test_recall = recall_score(all_labels, all_preds)
test_f1 = f1_score(all_labels, all_preds)

print(f"Test Loss: {avg_test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.2f}")
print(f"Test Precision: {test_precision:.2f}")
print(f"Test Recall: {test_recall:.2f}")
print(f"Test F1 Score: {test_f1:.2f}")

# Matrice de confusion
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Live', 'Fake'], yticklabels=['Live', 'Fake'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Courbe ROC et AUC
fpr, tpr, _ = roc_curve(all_labels, all_preds)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

```

```python
#Hi_Scan
import os
import numpy as np
from PIL import Image, UnidentifiedImageError
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import timm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 32
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_DIR = "LivDet2015/Testing"
SENSOR_NAME = "Hi_Scan"

# Transformations des images
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# Dataset personnalis
class LivDetBinaryDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        sensor_path = os.path.join(root_dir, SENSOR_NAME)
        for label in ['Live', 'Fake']:
            label_path = os.path.join(sensor_path, label)
            if not os.path.isdir(label_path):
                continue
            if label == 'Live':
                for img_file in os.listdir(label_path):
                    img_path = os.path.join(label_path, img_file)
                    if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                        self.samples.append((img_path, 0))
            else:
                for material in os.listdir(label_path):
                    material_path = os.path.join(label_path, material)
                    if not os.path.isdir(material_path):
                        continue
                    for img_file in os.listdir(material_path):
                        img_path = os.path.join(material_path, img_file)
                        if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                            self.samples.append((img_path, 1))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            with Image.open(img_path) as img:
                img = img.convert('L')
                if self.transform:
                    img = self.transform(img)
                return img, label
        except (OSError, UnidentifiedImageError) as e:
            print(f"[Skipping] Image corrompue ou illisible: {img_path} ({e})")
            return self.__getitem__((idx + 1) % len(self.samples))  # skip vers l'image suivante

# Charge les donnes de test
test_dataset = LivDetBinaryDataset(DATA_DIR, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Charger le modle pralablement sauvegard
CHECKPOINT_PATH = "checkpoints/Hi_Scan/checkpoint.pth" 
checkpoint = torch.load(CHECKPOINT_PATH)
model = timm.create_model('convnext_tiny', pretrained=False, num_classes=2)
model.load_state_dict(checkpoint['model_state_dict'])
model.to(DEVICE)
model.eval()

# Variables pour le calcul des mtriques
all_labels = []
all_preds = []

# Test du modle
test_loss = 0
criterion = nn.CrossEntropyLoss()

with torch.no_grad():
    loop = tqdm(test_loader, desc="Testing")
    for inputs, labels in loop:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        test_loss += loss.item()

        _, preds = torch.max(outputs, 1)
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

        loop.set_postfix(loss=test_loss / len(test_loader))

# Calcul des mtriques
avg_test_loss = test_loss / len(test_loader)
test_acc = accuracy_score(all_labels, all_preds)
test_precision = precision_score(all_labels, all_preds)
test_recall = recall_score(all_labels, all_preds)
test_f1 = f1_score(all_labels, all_preds)

print(f"Test Loss: {avg_test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.2f}")
print(f"Test Precision: {test_precision:.2f}")
print(f"Test Recall: {test_recall:.2f}")
print(f"Test F1 Score: {test_f1:.2f}")

# Matrice de confusion
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Live', 'Fake'], yticklabels=['Live', 'Fake'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Courbe ROC et AUC
fpr, tpr, _ = roc_curve(all_labels, all_preds)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

```

```python
#Test de Hi_Scan sur les autres capteurs 
import os
import torch
import timm
import numpy as np
from PIL import Image, UnidentifiedImageError
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Capteurs  tester
sensors = ['CrossMatch', 'GreenBit', 'Hi_Scan', 'Digital_Persona']
base_ckpt_path = "checkpoints"
base_test_path = "LivDet2015/Testing"

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# Dataset personnalis pour test
class FingerprintDataset(Dataset):
    def __init__(self, root_dir):
        self.samples = []
        self.transform = transforms.Compose([
            transforms.Grayscale(num_output_channels=3),
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize([0.5]*3, [0.5]*3)
        ])
        for label_type in ['Live', 'Fake']:
            folder = os.path.join(root_dir, label_type)
            label = 0 if label_type == 'Live' else 1
            if os.path.exists(folder):
                for subdir, _, files in os.walk(folder):
                    for file in files:
                        if file.lower().endswith(('.bmp', '.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                            self.samples.append((os.path.join(subdir, file), label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        try:
            image = Image.open(path).convert('L')
            image = self.transform(image)
            return image, label
        except (OSError, UnidentifiedImageError):
            return self.__getitem__((idx + 1) % len(self.samples))


# Fonction pour tester un modle entran sur un capteur donn
def evaluate_cross_sensor(training_sensor):
    model_path = os.path.join(base_ckpt_path, training_sensor, "checkpoint.pth")
    if not os.path.exists(model_path):
        return f"Checkpoint not found for {training_sensor}"

    # Charger le modle sauvegard
    model = timm.create_model('convnext_tiny', pretrained=False, num_classes=2)
    checkpoint = torch.load(model_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device)
    model.eval()

    results = {}

    for test_sensor in sensors:
        test_dir = os.path.join(base_test_path, test_sensor)
        dataset = FingerprintDataset(test_dir)
        loader = DataLoader(dataset, batch_size=32, shuffle=False)

        all_preds, all_labels = [], []

        with torch.no_grad():
            for imgs, labels in loader:
                imgs, labels = imgs.to(device), labels.to(device)
                outputs = model(imgs)
                preds = torch.argmax(outputs, dim=1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        acc = accuracy_score(all_labels, all_preds)
        prec = precision_score(all_labels, all_preds)
        rec = recall_score(all_labels, all_preds)
        f1 = f1_score(all_labels, all_preds)
        cm = confusion_matrix(all_labels, all_preds)

        results[test_sensor] = {
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1_score": f1,
            "confusion_matrix": cm
        }

    return results


# Lancer l'valuation pour un modle entran sur CrossMatch
cross_sensor_results = evaluate_cross_sensor("Hi_Scan")
cross_sensor_results
```
