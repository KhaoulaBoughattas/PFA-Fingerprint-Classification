```python
import os

def count_images_in_directory(root_path):
    distribution = {}

    for subset in ['Training', 'Testing']:
        subset_path = os.path.join(root_path, subset)
        if not os.path.exists(subset_path):
            continue

        for sensor in os.listdir(subset_path):
            sensor_path = os.path.join(subset_path, sensor)
            if not os.path.isdir(sensor_path):
                continue

            for label in ['Live', 'Fake']:
                label_path = os.path.join(sensor_path, label)
                if not os.path.exists(label_path):
                    continue

                if label == 'Live':
                    num_images = sum(len(files) for _, _, files in os.walk(label_path))
                    distribution.setdefault(subset, {}).setdefault(sensor, {})['Live'] = num_images
                else:
                    for material in os.listdir(label_path):
                        material_path = os.path.join(label_path, material)
                        if os.path.isdir(material_path):
                            num_images = len([f for f in os.listdir(material_path) if os.path.isfile(os.path.join(material_path, f))])
                            distribution.setdefault(subset, {}).setdefault(sensor, {}).setdefault('Fake', {})[material] = num_images

    return distribution

def print_distribution(dist):
    for subset, sensors in dist.items():
        print(f"\nSubset: {subset}")
        for sensor, data in sensors.items():
            print(f"  Sensor: {sensor}")
            if 'Live' in data:
                print(f"    live: {data['Live']} images")
            if 'Fake' in data:
                print("    Fake:")
                for material, count in data['Fake'].items():
                    print(f"      {material}: {count} images")

# Replace with the actual path to your dataset
dataset_path = 'LivDet2015'
distribution = count_images_in_directory(dataset_path)
print_distribution(distribution)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import cv2

class FingerprintDataset(Dataset):
    def __init__(self, base_path, sensors, transform=None):
        self.base_path = base_path
        self.sensors = sensors
        self.transform = transform
        self.images = []
        self.labels = []

        # Charger les images et leurs labels (0 = Live, 1 = Fake)
        for sensor in self.sensors:
            for category in ["live", "fake"]:
                category_path = os.path.join(self.base_path, sensor, category)
                
                # Si "fake", aller chercher les sous-dossiers (BodyDouble, Ecoflex, etc.)
                if category == "fake":
                    subfolders = [f for f in os.listdir(category_path) if os.path.isdir(os.path.join(category_path, f))]
                    for subfolder in subfolders:
                        folder_path = os.path.join(category_path, subfolder)
                        self.load_images(folder_path, label=1)
                else:
                    self.load_images(category_path, label=0)

    def load_images(self, folder_path, label):
        for img_name in os.listdir(folder_path):
            if img_name.lower().endswith('.bmp'):
                img_path = os.path.join(folder_path, img_name)
                self.images.append(img_path)
                self.labels.append(label)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]

        # Charger l'image en niveaux de gris et convertir en RGB
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

        if self.transform:
            img = self.transform(img)

        return img, label

# Définition des transformations pour ConvNeXt
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),  # ConvNeXt attend du 224x224
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalisation standard
])

# Charger le dataset
train_dataset = FingerprintDataset(base_path="LivDet2015/Training", sensors=sensors, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Vérification : Afficher la taille du dataset
print(f"Nombre total d'images : {len(train_dataset)}")
```

```python
import os
import cv2
import numpy as np
import albumentations as A
from tqdm import tqdm

# Définition des augmentations
transform = A.Compose([
    A.Rotate(limit=10, p=0.5),  # Rotation modérée
    A.GaussianBlur(blur_limit=(3, 5), p=0.3),  # Flou léger
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),  # Luminosité & contraste
    A.ElasticTransform(alpha=1, sigma=50, alpha_affine=50, p=0.3),  # Déformation mineure
    A.GaussNoise(var_limit=(10, 50), p=0.3),  # Ajout de bruit aléatoire
])

#  Fonction pour appliquer l'augmentation en conservant la structure
def augment_dataset(input_dir, output_dir, num_augmentations=3):
    for root, _, files in os.walk(input_dir):
        for file in tqdm(files, desc=f"Processing {root}"):
            if file.endswith('.bmp'):  # Vérifie les fichiers BMP uniquement
                img_path = os.path.join(root, file)
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Lecture en niveaux de gris

                if img is None:
                    print(f" Erreur de lecture : {img_path}")
                    continue
                
                img_resized = cv2.resize(img, (256, 256))  # Redimensionnement standard
                
                # Création du dossier dans `data_augmentation` avec la même structure
                relative_path = os.path.relpath(root, input_dir)  
                output_subdir = os.path.join(output_dir, relative_path)
                os.makedirs(output_subdir, exist_ok=True)

                for i in range(num_augmentations):
                    augmented = transform(image=img_resized)['image']
                    aug_img_path = os.path.join(output_subdir, f"{os.path.splitext(file)[0]}_aug{i}.bmp")
                    cv2.imwrite(aug_img_path, augmented)

input_dir = "LivDet2015/"
output_dir = "data_augmentation/"
augment_dataset(input_dir, output_dir, num_augmentations=5)  # Génère 5 augmentations par image
```

```python
import os
import cv2
import numpy as np
import albumentations as A
from tqdm import tqdm
input_dir = "LivDet2015"
output_dir = "LivDet2015_augmented"

# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_dir, exist_ok=True)

# Data Augmentation Transforms (corrigé)
transform = A.Compose([
    A.GaussNoise(var_limit=(10, 50), p=0.5),  
    A.MotionBlur(blur_limit=3, p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.ElasticTransform(p=0.5),
])

for dataset_type in ["Training", "Testing"]:
    input_path = os.path.join(input_dir, dataset_type)
    output_path = os.path.join(output_dir, dataset_type)
    os.makedirs(output_path, exist_ok=True)

    if not os.path.exists(input_path):
        print(f" {input_path} introuvable, ignoré.")
        continue

    #  Lire chaque sous-dossier (CrossMatch, GreenBit...)
    for subfolder in os.listdir(input_path):
        subfolder_path = os.path.join(input_path, subfolder)
        subfolder_output_path = os.path.join(output_path, subfolder)

        if not os.path.isdir(subfolder_path):
            continue

        os.makedirs(subfolder_output_path, exist_ok=True)

        #  **Lire les sous-sous-dossiers (Fake/Ecoflex, Fake/Gelatine...)**
        for root, _, files in os.walk(subfolder_path):
            relative_path = os.path.relpath(root, input_path)
            output_subfolder = os.path.join(output_path, relative_path)
            os.makedirs(output_subfolder, exist_ok=True)

            for file in tqdm(files, desc=f" Processing {relative_path}"):
                if file.lower().endswith((".bmp", ".png")):  #  Maintenant, il prend BMP et PNG
                    img_path = os.path.join(root, file)
                    image = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

                    if image is None:
                        print(f" Erreur de lecture : {file}")
                        continue

                    img_base, ext = os.path.splitext(file)

                    #  Appliquer Data Augmentation (5 versions)
                    for i in range(5):
                        augmented = transform(image=image)["image"]
                        output_file = f"{img_base}_aug_{i}{ext}"  #  Conserve l'extension originale
                        output_img_path = os.path.join(output_subfolder, output_file)
                        cv2.imwrite(output_img_path, augmented)

print("**Data Augmentation Complète !** Vérifie LivDet2015_augmented ")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import models
from tqdm import tqdm
import os
```

```python
# Configuration
num_classes = 5
learning_rate = 0.0001
epochs = 200
batch_size = 32
checkpoint_path = "best_model.pth"

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Load ConvNeXt model
model = models.convnext_tiny(pretrained=True)
model.classifier[-1] = nn.Linear(model.classifier[-1].in_features, num_classes)
model = model.to(device)
```

```python
# Loss function & Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5, verbose=True)

# Early stopping variables
best_val_loss = float('inf')
patience_counter = 0
```

```python
# Training Loop
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    
    train_loss = running_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}")
    
    # Validation
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
    val_loss /= len(val_loader)
    print(f"Validation Loss: {val_loss:.4f}")
    
    scheduler.step(val_loss)
    
    # Checkpointing
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), checkpoint_path)
        print(" Model saved!")
    else:
        patience_counter += 1
    
    # Early stopping
    if patience_counter >= patience:
        print(" Early stopping triggered!")
        break

print("Training completed! ")
```

```python
# 📦 Imports
import os
import torch
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurations
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
batch_size = 32
checkpoint_path = "best_model.pth"

#  Transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

#  Fonctions de chargement
def load_dataloader(path):
    dataset = datasets.ImageFolder(path, transform=transform)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)
    return loader, dataset.classes

#  Fonction de test
def test_model(model, loader, class_names, title=""):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    print(f"\n Résultats - {title}")
    print(classification_report(y_true, y_pred, target_names=class_names))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap="Blues")
    plt.title(f"Matrice de confusion - {title}")
    plt.xlabel("Prédictions")
    plt.ylabel("Réel")
    plt.show()
```

```python
# InterSensor (même capteur)
def run_intersensor(sensor="CrossMatch"):
    print(f"\n InterSensor | Capteur : {sensor}")
    test_path = f"LivDet2015_augmented/testing/{sensor}"
    test_loader, class_names = load_dataloader(test_path)

    model = models.convnext_tiny(weights=None)
    model.classifier[-1] = torch.nn.Linear(model.classifier[-1].in_features, 2)
    model.load_state_dict(torch.load(checkpoint_path))
    model.to(device)

    test_model(model, test_loader, class_names, title=f"InterSensor - {sensor}")
```

```python
# CrossSensor (capteur différent)
def run_crosssensor(train_sensor="CrossMatch", test_sensor="GreenBit"):
    print(f"\n CrossSensor | Train: {train_sensor} | Test: {test_sensor}")
    test_path = f"LivDet2015_augmented/testing/{test_sensor}"
    test_loader, class_names = load_dataloader(test_path)

    model = models.convnext_tiny(weights=None)
    model.classifier[-1] = torch.nn.Linear(model.classifier[-1].in_features, 2)
    model.load_state_dict(torch.load(checkpoint_path))
    model.to(device)

    test_model(model, test_loader, class_names, title=f"CrossSensor - Train: {train_sensor} / Test: {test_sensor}")

```

```python
import os
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from torchvision import models
import torch.nn as nn
from sklearn.metrics import accuracy_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
testing_root = "LivDet2015_augmented/Testing"
checkpoint_path = "best_model.pth"  # modèle entraîné sur 2 classes

# ------------------------
# Custom Dataset
# ------------------------
class FingerprintDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        self._prepare_dataset(root_dir)

    def _prepare_dataset(self, root_dir):
        for class_type in ['Live', 'Fake']:
            class_path = os.path.join(root_dir, class_type)
            if not os.path.isdir(class_path): continue

            label = 1 if class_type == 'Live' else 0

            if class_type == 'Fake':
                for material in os.listdir(class_path):
                    material_path = os.path.join(class_path, material)
                    for img_file in os.listdir(material_path):
                        self.samples.append((os.path.join(material_path, img_file), label))
            else:
                for img_file in os.listdir(class_path):
                    self.samples.append((os.path.join(class_path, img_file), label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, label

# ------------------------
# Transforms
# ------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# ------------------------
# Model Definition
# ------------------------
model = models.convnext_tiny(weights=None)
model.classifier[-1] = nn.Linear(model.classifier[-1].in_features, 2)

# Charger les poids
try:
    model.load_state_dict(torch.load(checkpoint_path, map_location=device))
except RuntimeError as e:
    print("Erreur de chargement :", e)
    exit()

model = model.to(device)
model.eval()

# ------------------------
# Évaluation par capteur
# ------------------------
capteur_scores = {}

for capteur in os.listdir(testing_root):
    capteur_path = os.path.join(testing_root, capteur)
    if not os.path.isdir(capteur_path): continue

    print(f" Évaluation sur le capteur : {capteur}")

    dataset = FingerprintDataset(capteur_path, transform=transform)
    dataloader = DataLoader(dataset, batch_size=32, shuffle=False)

    y_true = []
    y_pred = []

    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            y_true.extend(labels.numpy())
            y_pred.extend(predicted.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    capteur_scores[capteur] = acc
    print(f"✅ Précision sur {capteur} : {acc*100:.2f}%\n")

# ------------------------
# Résumé final
# ------------------------
print("Résultats globaux par capteur :")
for capteur, acc in capteur_scores.items():
    print(f" {capteur}: {acc*100:.2f}%")
```

```python
def test_on_sensor(sensor_name, base_dir="LivDet2015_augmented/Testing"):
    print(f"\n Test sur capteur : {sensor_name}")
    
    test_path = os.path.join(base_dir, sensor_name)
    dataset = datasets.ImageFolder(test_path, transform=transform)
    test_loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False)

    model = load_model()

    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    #  Regrouper en Live (classe 'Live') vs Fake (toutes les autres classes)
    idx_live = dataset.class_to_idx.get("Live", None)
    binary_preds = [0 if p == idx_live else 1 for p in all_preds]
    binary_labels = [0 if l == idx_live else 1 for l in all_labels]

    #  Résultats
    print(" Résultats (Live vs Fake):")
    print(f"Accuracy  : {accuracy_score(binary_labels, binary_preds):.4f}")
    print(f"Precision : {precision_score(binary_labels, binary_preds, zero_division=0):.4f}")
    print(f"Recall    : {recall_score(binary_labels, binary_preds, zero_division=0):.4f}")
    print(f"F1-score  : {f1_score(binary_labels, binary_preds, zero_division=0):.4f}")
    
    print("\nDetailed Report:")
    print(classification_report(binary_labels, binary_preds, target_names=["Live", "Fake"], zero_division=0))
    
    # Matrice de confusion
    cm = confusion_matrix(binary_labels, binary_preds)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Live", "Fake"], yticklabels=["Live", "Fake"])
    plt.title(f"Confusion Matrix - {sensor_name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```

```python
capteurs = ["CrossMatch", "Digital_Persona", "GreenBit", "Hi_Scan"]

for capteur in capteurs:
    test_on_sensor(sensor_name=capteur)
```
